````python
from client.models import LLM4PP_Problem, LLM4PP_Submission
from client.pareval_client import ParEvalDriver
from fastcoder.chatapi import MessageHistory, ChatAPI
import json
import os


ststem_prompt='''
You are a coding expert that writes very fast code.
You will be provided with a copy of the cpp code, which is a function.
You need to generate a compile-run cpp code, use the code to randomly generate test cases, and evaluate and output the running time of the function.
And you need to generate the corresponding compilation instructions.
The user will provide you a JSON dictionary in the following format:
```json
{
    "source_code" : <Initial code>
}
```
You need to follow exactly the following format to generate a response:
# Response format
```json
{
    "total_code" : <Generated code>,
    "compile_command" : <Compilation command>
}
```

'''

os.environ['OPENAI_API_KEY'] = ''
driver = ParEvalDriver()

chatAPI = ChatAPI(api_key='', base_url='https://api.openai.com/v1')

for problem in driver:
    problem : LLM4PP_Problem
    if problem.problem_id != "02_dense_la_gemm":
        continue
    messages = MessageHistory()
    messages.add_message("system", ststem_prompt)
    messages.add_message("user", json.dumps({"source_code.cpp": problem.source_code}))
    response = chatAPI.get_response('gpt-4o', messages, json_format=True)
    optimized_code = json.loads(response)['total_code']
    compile_command = json.loads(response)['compile_command']
    
    print(optimized_code)
    print(compile_command)
````
